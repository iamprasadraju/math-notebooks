**what is a tangent ?** [Def](https://en.wikipedia.org/wiki/Tangent)

The tangent line (or simply tangent) to a plane curve at a given point is, intuitively, the straight line that "just touches" the curve at that point.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/0/0f/Tangent_to_a_curve.svg/960px-Tangent_to_a_curve.svg.png" width="25%">

- **slope**: change in y over change in x. [Def](https://en.wikipedia.org/wiki/Slope)

  when points $(x, y) , (x_0, y_0)$

  $m = \frac{\Delta y}{\Delta x}$ i.e $m = \frac{y_2 - y_1}{x_2 - x_1}$


  *Point-slope form* :
  - multiply both sides with $(x - x_0)$:

    $y - y_1 = m(x - x_0)\rightarrow$  *line eq*

  *Two-point form* :

  $$y - y_1 = \frac{y_2 - y_1}{x_2 - x_1} (x - x_1)$$
