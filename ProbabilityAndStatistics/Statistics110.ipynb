## Probability
---

Probability measures the likelihood of an event occurring, expressed as a number between 0 (impossible) and 1 (certain), or as a percentage.

<span style="background-color: yellow; font-weight: bold" >Definition:</span> Probability concerns events and numerical descriptions of how likely they are to occur. The probability of an event is a number between 0 and 1; the larger the probability, the more likely an event is to occur.This number is often expressed as a percentage (%), ranging from 0% to 100%. [Wiki](https://en.wikipedia.org/wiki/Probability). 
$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}
$$

$$\text{where:}
\begin{cases}
P(A) & \text{is the probability of event } A \\
\text{Favorable outcomes} & \text{are the outcomes in which event } A \text{ occurs} \\
\text{Total outcomes} & \text{are all possible outcomes(Sample space)}
\end{cases} $$

<u>Assumptions (Classical Probability):</u>
1. All outcomes equally likely
2. Finite sample space 

- **Sample Space**: A set of all possible outcomes of an experiment.
- **Event**: An event is a subset of sample space.




### How to Count ?
---

**Counting**: Counting is the process of determining the number of elements of a finite set of objects; that is, determining the size of a set.

<span style="background-color: yellow">Multiplication Rule</span> (Rule of product): if have an experiment with $n_1$ possible outcomes, and for each outcome of 1st experiment there are $n_2$ outcomes for 2nd experiment, ...... , for each their are $n_r$ outcomes for rth experiment, then $n_1 \times n_2 ..... \times n_r$ overall possible outcomes.[Wiki](https://en.wikipedia.org/wiki/Rule_of_product)

**Binomial coefficient**$\binom {n}{k}$: the number of ways to choose a subset of $k$ elements from a larger set of $n$ distinct elements, where order does not matter. (combinations) [Wiki](https://en.wikipedia.org/wiki/Binomial_coefficient)

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!} \quad \text{if } k \leq n, \quad \text{and } 0 \text{ if } k > n
$$
