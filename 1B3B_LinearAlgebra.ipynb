# Linear Algebra

 



```python
import matplotlib.pyplot as plt
import random

# random.seed(42)

O = [0, 0]

A = [random.randint(1, 12) for _ in range(2)]

B = [random.randint(1, 12) for _ in range(2)]


print(A, B)
```

```python
# Vector addition
def vector_add(p1, p2):
    if len(p1) == len(p2):
        return [p1[i] + p2[i] for i in range(len(p1))]


# Vector difference
def vector_sub(p1, p2):
    if len(p1) == len(p2):
        return [p1[i] - p2[i] for i in range(len(p1))]

result = vector_add(A, B)

def plot_2dvectors(vectors, origins):
    fig, ax = plt.subplots(figsize = (6, 6)) # Initialize size

    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    
    # 3. Hide the default "top" and "right" spines
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    # 4. Ensure ticks are displayed on the new bottom and left axes
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    for vector, origin in zip(vectors, origins):
        ax.quiver(origin[0], origin[1], vector[0], vector[1], scale_units='xy', angles='xy', scale=1)
    
    ax.set_xlim([-30, 30])
    ax.set_ylim([-30, 30])
    
    ax.grid(True, linestyle='--', alpha=0.6)
    # ax.legend()
    
    plt.show() # show plot


plot_2dvectors(vectors = [A, B, result] , origins = [O, O, O])
```
