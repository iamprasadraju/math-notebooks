# Linear Algebra

 



```python
import matplotlib.pyplot as plt
import random

O = [0, 0]

A = [random.randint(1, 9) for _ in range(2)]

B = [random.randint(1, 9) for _ in range(2)]
```

```python
def vector_add(p1, p2):
    if len(p1) == len(p2):
        return [p1[i] + p2[i] for i in range(len(p1))]

def vector_sub(p1, p2):
    if len(p1) == len(p2):
        return [p1[i] - p2[i] for i in range(len(p1))]

result = vector_sub(A, B)

def plot_2dvectors(vectors, origins):
    fig, ax = plt.subplots(figsize = (8, 8)) # Initialize size

    for vector, origin in zip(vectors, origins):
        ax.quiver(origin[0], origin[1], vector[0], vector[1], scale_units='xy', angles='xy', scale=1)
    
    ax.set_xlim(-1, 2 * max(result))
    ax.set_ylim(-1, 2 * max(result))
    
    ax.grid(True) # Grid layout
    
    plt.show() # show plot


plot_2dvectors(vectors = [A, B , result] , origins = [O, O, O])
```
